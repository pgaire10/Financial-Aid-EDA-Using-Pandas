```python
#1 Importing Pandas and asssignig the pandas as pd
import pandas as pd
```

```python
#2 Loading CSV file into the dataframe.
#Read the file named grades
df1 = pd.read_csv('grades.csv')
```

```python
#3 Viewing the first 10 rows of the data from the dataframe df1
print(df1.head(10))
```

```python
#4 Viewing the bottom 10 rows of the data from the dataframe df1
print(df1.tail(10))
```

```python
#5 Displaying the basic summary statistics from the dataset grades
df1.describe()
```

```python
#6 Checking for number of duplicate rows in the dataframe
duplicate_count = df1.duplicated().sum()
print(duplicate_count)

#Displaying the dupicate rows if there are any of them
duplicated_rows = df1[df1.duplicated()]
duplicated_rows
```

```python
#7 Droping those 4 duplicate rows from the dataset
df1.drop_duplicates(inplace=True)
```

```python
#8 Checking for the missing values in each rows
print(df1.isnull().sum())

#Droping those missing values rows from the dataset
df1.dropna(inplace=True)
print(df1.isnull().sum())
```

```python
#9 Creating a column named “Honor Student” and putting a “Yes” if their GPA exceeds 3.5 and a “No” if it doesn’t
def honor_student(gpa):
    if gpa > 3.5:
        return "Yes"
    else:
        return "No"
df1['Honor Student'] = df1['GPA'].map(honor_student)

#Viewing the dataframe with the column "Honor Student" 
print(df1)
```

```python
#10 Loading an excel file named “tuition” into a dataframe called “df2”
df2 = pd.read_excel('tuition.xlsx')
```

```python
#11 Checking dataframe 2 i.e. df2 for duplicate rows if they exist
duplicate_count = df2.duplicated().sum()
print(duplicate_count)

#Displaying the dupicate rows if there are any of them
duplicated_rows = df2[df2.duplicated()]
duplicated_rows
```

```python
#12 Droping that one duplicate row from the dataframe 2
df2.drop_duplicates(inplace=True)
```

```python
#13 Checking the missing values in each rows from the dataframe 2
print(df2.isnull().sum())

#Droping those missing values rows from the dataframe 2
df2.dropna(inplace=True)
print(df2.isnull().sum())
```

```python
#14 Creating a column called “Cost” that is the result of subtracting financial aid from tuition.
df2['Cost'] = df2['Tuition'] - df2['Financial Aid']

#Viewing the dataframe with the new column "Cost"
print(df2)
```

```python
#15 Merging the two dataframes using the full join on the "ID" Column
df3 = pd.merge(df1, df2, on="ID", how="outer")

#Displaying the merged dataframe
print(df3)
```

```python
#16 Checking for the  missing values and droping the rows
print(df3.isnull().sum())

df3.dropna(inplace=True)
print(df3.isnull().sum())
```

```python
#17 Grouping the students by major and displaying the mean GPA and mean cost
df3.groupby(['Major']).aggregate({'GPA':['mean'],'Cost':['mean']})
```

```python
#18 Filtering everyone with GPA below 2.0 from the dataframe df3
df3[df3['GPA'] > 2.0]
```

```python
#19 Sorting by ID in a descending order
ID_Desc = df3.sort_values('ID', ascending=False)
print(ID_Desc)
```

```python
#20 Exporting df3 as a CSV file
df3.to_csv('output.csv',index=False)
```

```python
#Exporting df3 as an Excel file
df3.to_excel('output.xlsx',index=False)
```
